### Text Preprocessing Strategy
***

1) convert to lowercase
2) delete special characters (symbols and hyphens)
3) possessive pronouns: for example reduce student's to student
4) delete double white spaces
5) remove dates and numbers
6) remove stop words
7) lemmatization instead of stemming: find root of word

- for tokenization we simply split by whitespaces with nltk

```python
def preprocess_text(text:str=None):
    """function to preprocess string
    returns a list of tokens
    """
    #1 to lowercase
    text = text.lower()
    
    #2 remove all special characters
    text = re.sub(r"\W", " ", text)
    
    #3 remove single characters with space to the left and right (possessive pronoun)
    text = re.sub(r"\s+[a-z]\s+", " ", text)
    
    #4 remove double whitespaces to single space
    text = re.sub(' +', ' ', text)
    
    #5.1 replace dates (human readable)
    # matches = search_dates(text)
    # if matches is not None:
    #     for match in matches:
    #         text = text.replace(str(match[0]), '<dates>')
    
    #5.2 replace numbers
    text = re.sub(r'\d+', '<num>', text)
    
    #5.3 now count frequncy of <dates> and <num>
    count_num, count_dates = text.count("<num>"), text.count("<dates>")
    
    #5.4 replace it (so that in tokens it does not appear)
    text = text.replace("<num>", "").replace("<dates>", "")
    
    #6 Stop words and tokenization
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    token_text = [i for i in tokens if not i in stop_words]
    
    #7 Lemmatization
    stemmer = WordNetLemmatizer()
    result = [stemmer.lemmatize(word) for word in token_text]
    
    return result
```

```python
from tqdm import tqdm

def merge_tokens(data:list):
    """
    function that preprocesses text and returns two different lists
    1. token_list ... used for dictionary to find out threshold value
    2. documents ... used for removing oov words
    """
    i = 0
    token_list = list() # list with all tokens of all douments ["word1", "word2", ...]
    documents = list() # elements are list of tokens [[doc1 tokens], [doc2 tokens], ...]
    
    for sample in tqdm(data):
        token_list += preprocess_text(sample[1])
        documents.append(preprocess_text(sample[1]))
        i += 1

        
    return token_list, documents
```

### Dictionary of vocabularies
***
- Goal: find out ***cutoff threshold*** by

1) Merging all text excerpts in one list
2) create dictionary
3) look at frequency distribution of word occurences
4) find threshold
5) get list of Out-of-vocabulary words to remove from text

```python
def create_dictionary(preprocessed_tokens:list=None):
    """Creates a word dictionary given a PREPROCESSED text. Returns a sorted dct of all counts and OOV as well as the count list"""
    threshold = 2

    count_list = list()
    filtered_dict = dict()
    out_of_vocabulary = list()
    
    word_dict = defaultdict(lambda: 0)
    for word in preprocessed_tokens:
        word_dict[word] += 1
    
    sorted_dict = {k: v for k, v in sorted(word_dict.items(), key=lambda item:item[1], reverse=True)}
    
    for key, value in sorted_dict.items():
        if value > threshold:
            filtered_dict[key] = value
        else:
            out_of_vocabulary.append(key)

        count_list.append(value)
    
    return sorted_dict, out_of_vocabulary, count_list
```

```python
train_tokens, train_documents = merge_tokens(train) #tokens, texts
val_tokens, val_documents = merge_tokens(validation)
test_tokens, test_documents = merge_tokens(test)
```

```python
train_dict, train_oov, train_count = create_dictionary(train_tokens)
val_dict, val_oov, val_count = create_dictionary(val_tokens)
test_dict, test_oov, test_count = create_dictionary(test_tokens)
```

### Out of vocabulary tokens
***

- We look at the frequency distribution of values and see that words that appear once in the dictionary make about 40% and words that appear twice occur even 3000 times so we set the ***threshold to 2*** and obtain a dictionary with about 8000 entries

```python
def oov_remover(oov:list,data_document:list):
    """First this function removes OOV and
    secondly since we deal with a list of list of strings
    we convert it to a list of strings for the CountVectorizer
    returns list of strings [doc1, doc2, doc3, ...]"""
    
    for oov in tqdm(oov):
        for doc in data_document:
            if oov in doc:
                doc.remove(f"{oov}")
                
    for idx, elem in enumerate(data_document):
        #data_document[idx] = " ".join(elem)
        data_document[idx] = elem
    
    return data_document
```

```python
train_documents2 = oov_remover(train_oov, train_documents)
val_documents2 = oov_remover(val_oov, val_documents)
test_documents2 = oov_remover(test_oov, test_documents)
```

<h3 style="color:rgb(0,130,170)">OOB </h3>
With this we remove extramly freqent words. We create a dictionary by removing extremely frequent words and keeping the top MAX_NB_WORDS=X words.<br><br>

After that, we create feature Vectors of Documents with dic:

```python
MAX_NB_WORDS = 1500

dic = Dictionary(train_documents2)
dic.filter_extremes(no_below=1, no_above=0.5, keep_n=MAX_NB_WORDS)
dic.compactify()
print("Total %s unique tokens." % len(dic))

docs_train_bow = [dic.doc2bow(_doc) for _doc in train_documents] 
docs_val_bow = [dic.doc2bow(_doc) for _doc in val_documents] 
docs_test_bow = [dic.doc2bow(_doc) for _doc in test_documents] 
```

<h3 style="color:rgb(0,130,170)">Creating 2 vector represantations (TF and TF-IDF) vectors</h3>
To create the the differnt vectors, we train the TfidfModel model using the bag-of-word representations of training data.

#### First, create general sentence Vectors

```python
weight_model = TfidfModel(docs_train_bow)

docs_train_tfidf=[]
docs_val_tfidf=[]
docs_test_tfidf=[]

for _bow in docs_train_bow:
    docs_train_tfidf.append(weight_model[_bow])
for _bow in docs_val_bow:
    docs_val_tfidf.append(weight_model[_bow])
for _bow in docs_test_bow:
    docs_test_tfidf.append(weight_model[_bow])
```

#### Now create sparse matrices in order to apply SVD later

```python
# create sparce tf matrix
docs_train_tf_sparse = np.zeros((len(docs_train_bow), MAX_NB_WORDS))
for doc_i, doc in enumerate(docs_train_bow):
    for item in doc:
        docs_train_tf_sparse[doc_i][item[0]] = np.log(1+ item[1])

docs_val_tf_sparse = np.zeros((len(docs_val_bow), MAX_NB_WORDS))
for doc_i, doc in enumerate(docs_val_bow):
    for item in doc:
        docs_val_tf_sparse[doc_i][item[0]] = np.log(1+ item[1])

docs_test_tf_sparse = np.zeros((len(docs_test_bow), MAX_NB_WORDS))
for doc_i, doc in enumerate(docs_test_bow):
    for item in doc:
        docs_test_tf_sparse[doc_i][item[0]] = np.log(1+ item[1])

# create sparce tfidf matrix 
docs_train_tfidf_sparse = np.zeros((len(docs_train_tfidf), MAX_NB_WORDS))
for doc_i, doc in enumerate(docs_train_tfidf):
    for item in doc:
        docs_train_tfidf_sparse[doc_i][item[0]] = item[1]
        
docs_val_tfidf_sparse = np.zeros((len(docs_val_tfidf), MAX_NB_WORDS))
for doc_i, doc in enumerate(docs_val_tfidf):
    for item in doc:
        docs_val_tfidf_sparse[doc_i][item[0]] = item[1]
        
docs_test_tfidf_sparse = np.zeros((len(docs_test_tfidf), MAX_NB_WORDS))
for doc_i, doc in enumerate(docs_test_tfidf):
    for item in doc:
        docs_test_tfidf_sparse[doc_i][item[0]] = item[1]

sparsity_train = 1.0 - (np.count_nonzero(docs_train_tfidf_sparse) / float(docs_train_tfidf_sparse.size))
print("Sparisty of the tfidf before SVD:", sparsity_train)
```

#### Finally apply TF, TF-IDF

```python
k = 150 # choose reasonable number

svd = TruncatedSVD(n_components=k)
svd.fit(docs_train_tfidf_sparse)

vecs_train_tf = svd.transform(docs_train_tf_sparse)
vecs_val_tf = svd.transform(docs_val_tf_sparse)
vecs_test_tf = svd.transform(docs_test_tf_sparse)

vecs_train_tfidf = svd.transform(docs_train_tfidf_sparse)
vecs_val_tfidf = svd.transform(docs_val_tfidf_sparse)
vecs_test_tfidf = svd.transform(docs_test_tfidf_sparse)
```

```python
# for rating the sparsitys (later)

def sparsity_rate(array):
    sparsity_rate = 1 - (np.count_nonzero(array) / float(array.size))
    return sparsity_rate
```

#### LSA

```python
k = 200 # choose reasonable number

svd = TruncatedSVD(n_components=k)
svd.fit(docs_train_tfidf_sparse)

vecs_train = svd.transform(docs_train_tfidf_sparse)
vecs_val = svd.transform(docs_val_tfidf_sparse)
vecs_test = svd.transform(docs_test_tfidf_sparse)
```

```python
print("Sparisty of the tf after SVD:", sparsity_rate(vecs_train_tf))
print("Sparisty of the tfidf after SVD:", sparsity_rate(vecs_train_tfidf))

print(f'The shape of the sparse matrix is: {docs_train_tfidf_sparse.shape}')
print(f'The shape of the lsa matrix is: {vecs_train_tfidf.shape}')
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay

def plot_classification_report(classificationReport,
                               title='Classification report',
                               cmap='RdBu'):

    classificationReport = classificationReport.replace('\n\n', '\n')
    classificationReport = classificationReport.replace(' / ', '/')
    lines = classificationReport.split('\n')

    classes, plotMat, support, class_names = [], [], [], []
    for line in lines[1:]:  # if you don't want avg/total result, then change [1:] into [1:-1]
        line = line.replace("macro avg", "mavg")
        line = line.replace("weighted avg", "wavg")
        if "accuracy" in line:
            continue
            
        t = line.strip().split()
        if len(t) < 2:
            continue
        classes.append(t[0])
        v = [float(x) for x in t[1: len(t) - 1]]
        support.append(int(t[-1]))
        class_names.append(t[0])
        plotMat.append(v)

    plotMat = np.array(plotMat)
    plt.rcParams["figure.figsize"] = (25,8)

    xticklabels = ['Precision', 'Recall', 'F1-score']
    yticklabels = ['{0} ({1})'.format(class_names[idx], sup)
                   for idx, sup in enumerate(support)]

    plt.imshow(plotMat)
    plt.title(title)
    #plt.colorbar()
    plt.xticks(np.arange(3), xticklabels, rotation=45)
    plt.yticks(np.arange(len(classes)), yticklabels)

    upper_thresh = plotMat.min() + (plotMat.max() - plotMat.min()) / 10 * 8
    lower_thresh = plotMat.min() + (plotMat.max() - plotMat.min()) / 10 * 2
    for i, j in itertools.product(range(plotMat.shape[0]), range(plotMat.shape[1])):
        plt.text(j, i, format(plotMat[i, j], '.2f'),
                 horizontalalignment="center",
                 color="white" if (plotMat[i, j] > upper_thresh or plotMat[i, j] < lower_thresh) else "black")

    plt.ylabel('Metrics')
    plt.xlabel('Classes')
    plt.tight_layout()
    return plt
```

Train a logistic regression model on the high-dimensional vectors. Take the coefficient weights, learned by the model, on each dimension (which here corresponds to each token in the dictionary). Separately for each class, study what are the tokens that have the highest contributions/importance for the predictions of the model.

```python
from sklearn.metrics import mean_squared_error
from sklearn.datasets import make_friedman1
from sklearn.ensemble import GradientBoostingRegressor

n_estimators=100
est = GradientBoostingRegressor(n_estimators=n_estimators, learning_rate=0.1, max_depth=1, random_state=0,
                                loss='ls').fit(vecs_train_tf, y_train)

mean_squared_error(y_test, est.predict(vecs_test_tf))
```

```python
y_pred = est.predict(vecs_test_tf)
test_score = np.zeros((n_estimators,), dtype=np.float64)

for i, y_pred in enumerate(est.staged_predict(vecs_test_tf)):
    test_score[i] = est.loss_(y_test, y_pred)

fig = plt.figure(figsize=(6, 6))
plt.subplot(1, 1, 1)
plt.title("Deviance")
plt.plot(
    np.arange(n_estimators) + 1,
    est.train_score_,
    "b-",
    label="Training Set Deviance",
)
plt.plot(
    np.arange(n_estimators) + 1, test_score, "r-", label="Test Set Deviance"
)
plt.legend(loc="upper right")
plt.xlabel("Boosting Iterations")
plt.ylabel("Deviance")
fig.tight_layout()
plt.show()
```
